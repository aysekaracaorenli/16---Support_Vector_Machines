# Support Vector Machines with Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
%matplotlib inline
```

## Get the Data

```python
df = pd.read_table('diabetesdata.txt')
y = df['Diabetes']
X = df[['BMI','Glucose']]
X
```

## Train Test Split and Standardize

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=5,stratify=y)
scaler = preprocessing.StandardScaler()
scaler.fit(X_train)
X_trainStandard = scaler.transform(X_train)
X_testStandard = scaler.transform(X_test)
```

# Train the Support Vector Classifier

```python
from sklearn.svm import SVC
```

```python
SVC()
```

```python
model = SVC(C=1,kernel='linear')
model.fit(X_trainStandard,y_train)
```

## Predictions and Evaluations

Now let's predict using the trained model.

```python
predictions = model.predict(X_testStandard)
model.score(X_testStandard,y_test)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
#Functions to plot decision boundaries
def make_meshgrid(x, y, h=.02):
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    return xx, yy

def plot_contours(ax, clf, xx, yy, **params):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
```

```python
fig, ax = plt.subplots()
# title for the plots
title = ('Decision surface of SVC ')
# Set-up grid for plotting.
X0, X1 = X_trainStandard[:, 0], X_trainStandard[:, 1]
xx, yy = make_meshgrid(X0, X1)

plot_contours(ax, model, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
ax.scatter(X0, X1, c=y_train, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
ax.set_ylabel('Glucose')
ax.set_xlabel('BMI')
ax.set_xticks(())
ax.set_yticks(())
ax.set_title(title)
plt.show()
```

# With all variables

```python
df = pd.read_table('diabetesdata.txt')
X = df.drop('Diabetes',axis=1)
y = df['Diabetes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=35,stratify=y)
scaler = preprocessing.StandardScaler()
scaler.fit(X_train)
X_trainStandard = scaler.transform(X_train)
X_testStandard = scaler.transform(X_test)
```

```python
model = SVC(C=1,kernel='linear')
model.fit(X_trainStandard,y_train)
```

```python
predictions = model.predict(X_testStandard)
model.score(X_testStandard,y_test)
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
pipe=Pipeline([('scaler',preprocessing.StandardScaler()),('Classifier',SVC())])
tuned_parameters = [{'Classifier__kernel': ['rbf'], 
                     'Classifier__gamma': [1,0.1,0.01,0.001],
                     'Classifier__C': [1, 10, 100, 1000]},
                    {'Classifier__kernel': ['linear'], 'Classifier__C': [1, 10, 100, 1000]},
                   {'Classifier__kernel':['poly'], 'Classifier__C':[1, 10, 100, 1000]}]
grid = GridSearchCV(pipe,tuned_parameters,refit=True,verbose=2,scoring='accuracy')
# May take awhile!
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
```

```python
model = SVC(C=10,gamma=0.001,kernel='rbf')
model.fit(X_trainStandard,y_train)
predictions = model.predict(X_testStandard)
```

```python
model.score(X_testStandard,y_test)
```

```python
print(confusion_matrix(y_test,predictions))
```
